## <h1 style="color: green;">3. Explanation of algorithms: </h1>

A* algorithm is used to calculate the shortest distance between the source(initial state) and the destination (final state).
A* algorithm has 3 parameters:
<br>**- g :** the cost of moving from the initial cell to the current cell. Basically,it is the sum of all the cells that have been visited since leaving the first cell.
<br>**- h :** also known as the heuristic value, it is the estimated cost of movingfrom the current cell to the final cell. The actual cost cannot becalculated until the final cell is reached. Hence, h is the estimated cost.We must make sure that there is never an over estimation of the cost.
<br> **- f :** it is the sum of g and h. So, $f = g + h$
<br> The way that the algorithm makes its decisions is by taking the f-valueinto account. The algorithm selects the smallest f-valued cell and movesto that cell. This process continues until the algorithm reaches its goalcell.

<br>-g: تكلفة الانتقال من الخلية الأولية إلى الخلية الحالية. في الأساس ، هو مجموع جميع
الخلايا التي تمت زيارتها منذ مغادرة الخلية الأولى.

## <h1 style="color: green;">3.1. Algorithm Implementation : </h1>

 <br> **Step 1 :** build Graph class that contains vertex and  A* method . 
<br> **Step 2 :** A* method has two sets *expands* and *explored* ,the first one is for nodes that are open but still not examined ; the other one is for nodes that are has been examined  .
<br> **Step 3 :** : initially , place *Client* to expands set and determain its $ f(n)=0$ .
<br> **Step 4 :** : remove *client* from the set and place it in explored set , and find its successors and add them to expands set and find their $ f(n) $ value .
<br> **Step 5 :** remove the minimum $ f(n) $ value from the set expands and add it to explored , check if it is *Server* then return the path , else and find its successors and add them to expands set and find their $ f(n) $ value .
<br> **Step 6 :** repeat step 5 untill expaneds set is empty .
    
<br>الخطوة الاولى : انشاء كلاس القراف والذي يحتوي على خوارزمية اي ستار .
<br> الخطوة الثانية : خوارزمية اي ستار تحتوي على مصفوفتين الاولى "اكسباند" والثانية "اكسبلورد"الاولى تحتوي على الاطراف التي فتحت والثانية تحتوي على الاطراف التي قمنا باختباها  
<br>الخطوة الثالة : نضع العميل بمصفوفة اكسباند ونجعل دالة المسافة لديه تساوي 0 
<br>الخطوة الرابعة : نزيل العميل من المصفوفة الاولى ونضعه بمصفوفة اكسبلورد ثم نوجد كل الابناء بالنسبة له ونضعهم بالمصفوفة الاولى مع ايجاد معادلة المسافة لهم
<br>الخطوة الخامسة : نزيل الطرف الذي لديه اقل مسافة من المصفوفة الاولى و نتحقق اذا كان هو الخادم ، فاذا كان الخادم نرجع المسار والا نوجد ابناءه ونضيفهم الى المصفوفة الاولى 
<br> الخطوة السادسة : نعيد الخطوة الخامسة الى ان تصبح المصفوفة الاولى فارغة 






```python
from collections import deque
 
class Graph:
    # adjac_lis , Heuristic will be set as properties in the Graph class
    # adjac_lis , في كلاس القراف Heuristic  سيتم تعيين خصائص ال   

    def __init__(self, adjac_lis , Heuristic):
        self.adjac_lis = adjac_lis
        self.Heuristic = Heuristic
        
    # get the neighbors of the node
    #الحصول على جميع النود المجاوره
    def get_neighbors(self, v):
        return self.adjac_lis[v]
 
    # This is function will find a node having equal values in the Heuristic
    #Heuristicهذه الداله سوف تجد النود التي لها قيم متساوية في ال
    def h(self, n):
        return self.Heuristic[n]
 
    def a_star_algorithm(self, start, goal):
        # expands is a list of visited nodes, not all of their neighbours inspected and it starts with the start node.
        # explored is a list of visited nodes, all of neighbours have been inspected and start with empty list
        #الاكسباند عبارة عن قائمة بالنود التي تمت زيارتها ، ولم يتم فحص جميع النود المجاوره ويبدأ بالنود الأولى.
        # الاكسبلور عبارة عن قائمة بالنود التي تمت زيارتها ، وقد تم فحص جميع النود المجاوره والبدء بقائمة فارغه
        
        expands = set([start])
        explored = set([])
 
        # Backword_des list of distances from start to all nodes it passes through and the default value is +infinity
        # Backword_des + ∞  قائمة المسافات من البداية إلى جميع النود التي تمر عبرها والقيمة الافتراضية هي
        Backword_des = {}
        Backword_des[start] = 0
 
        # adjacent_nodes list of contains adjacent nodes
        #adjacent_nodesقائمة تحتوي على النود المجاورة
        adjacent_nodes = {}
        adjacent_nodes[start] = start
        
        # use lan() To know the number of items in the list
        # لمعرفة عدد العناصر في القائمة lan ()  يتم استخدام 

        while len(expands) > 0:
            n = None
 
            # it will find a node with the lowest value 
            # سيتم ايجاد النود التي تكون تملك أقل قيمة
            for v in expands:
                if n == None or Backword_des[v] + self.h(v) < Backword_des[n] + self.h(n):
                    n = v;
 
            if n == None:
                print('Path does not exist!')
                return None
 
            # if the current node is the goal , then we start again from start
            #اذا كانت النود الحاليه هي القول ، فسنبدأ مرة أخرى من البداية
            if n == goal:
                reconst_path = []
                # fill reconst_path list with the previous path
                #Recst_path ملء قائمة  بالمسار السابق
                while adjacent_nodes[n] != n:
                    reconst_path.append(n)
                    n = adjacent_nodes[n]
 
                reconst_path.append(start)
                # reverse reconst_path list
                #reconst_path عكس القائمه
                reconst_path.reverse()
 
               #method to print the path is optimal or not 
                print('cost from Cient to the Server by using A* is ' , Backword_des[goal])
                if self.Shortest_path(start, goal,Backword_des[goal])== True :
                    print("A* path is Optimal ")
                else :
                    print("A* path is Not Optimal ")
                return reconst_path
 
            # for all the neighbors of the current node do
            # لجميع النود المجاوره للنود الحالية يفعلون الاتي 

            for (m, weight) in self.get_neighbors(n):
              # If the current node is not present in both expands and explored, add it to expands.
                # and current node will be in adjacent_nodes
                # إذا كانت النود الحالية غير موجودة في كل من الاكسباند و الاكسبلور ، فقم بإضافتها إلى الاكسباند.
                # وستكون النود الحالية في النود المجاوره
                
                if m not in expands and m not in explored:
                    expands.add(m)
                    adjacent_nodes[m] = n
                    Backword_des[m] = Backword_des[n] + weight
 
                # otherwise, check if it's quicker to first visit n, then m
                # and if it is, update adjacent_nodes data and Backword_des data
                # and if the node was in the explored, move it to expands
                # m اولا ثم n  خلاف ذلك ، تحقق ممن  كان  الأسرع  في الزيارة  
                # Backword_des and adjacent_nodes وإذا كان الأمر كذلك ، فقم بتحديث بيانات    
                # وإذا كانت النود في الاكسبلور ، فانقلها إلى الاكسباند
                
                
                else:
                    if Backword_des[m] > Backword_des[n] + weight:
                        Backword_des[m] = Backword_des[n] + weight
                        adjacent_nodes[m] = n
 
                        if m in explored:
                            explored.remove(m)
                            expands.add(m)
 
            # remove n from the expands, and add it to explored
            # because all of his neighbors were inspected
            #ازاله (ان ) من الاكسباند و اضافتها الى الاكسبلور
            #لأنه تم تفتيش جميع النود المجاوره
            
            expands.remove(n)
            explored.add(n)
 
        print('Path does not exist!')
        return None
    # here we use the same method of A* but we remove the h(n) to decide the path is optimal or nor
    #هنا نستخدم نفس طريقة A *
    #لكننا نزيل h (n) 
    #لنقرر أن المسار هو الأمثل أو لا
    
    
    def Shortest_path(self, start, goal,astar_path):
        # expands is a list of visited nodes, not all of their neighbours inspected and it starts with the start node.
        # explored is a list of visited nodes, all of neighbours have been inspected and start with empty list
        # الاكسباند عبارة عن قائمة بالنود التي تمت زيارتها ، ولم يتم فحص جميع النود المجاوره لها ويبدأ بالنود البداية
        #  الاكسبلور عبارة عن قائمة بالنودد التي تمت زيارتها ، وقد تم فحص   جميع النود المجاو والبدء بقائمة فارغة
        
        expands = set([start])
        explored = set([])
 
        # Backword_des list of distances from start to all nodes it passes through and the default value is +infinity
        # Backword_des + ∞  قائمة المسافات من البداية إلى جميع النود التي تمر عبرها والقيمة الافتراضية هي 

        Backword_des = {}
        Backword_des[start] = 0
 
        # adjacent_nodes list of contains adjacent nodes
        #adjacent_nodesقائمة تحتوي على النود المجاورة 

        adjacent_nodes = {}
        adjacent_nodes[start] = start
        
        # use lan() To know the number of items in the list
        # لمعرفة عدد العناصر في القائمة lan ()  يتم استخدام 

        while len(expands) > 0:
            n = None
 
            # it will find a node with the lowest value 
            # سيتم ايجاد النود التي تكون تملك أقل قيمة

            for v in expands:
                if n == None or Backword_des[v] < Backword_des[n] :
                    n = v;
 
            if n == None:
                print('Path does not exist!')
                return False
 
            # if the current node is the goal , then we start again from start
            #اذا كانت النود الحاليه هي القول ، فسنبدأ مرة أخرى من البداية 

            if n == goal:
                reconst_path = []
                # fill reconst_path list with the previous path
                # Recst_path ملء قائمة  بالمسار السابق

                while adjacent_nodes[n] != n:
                    reconst_path.append(n)
                    n = adjacent_nodes[n]
 
                reconst_path.append(start)
                # reverse reconst_path list
                #reconst_path عكس القائمه

                reconst_path.reverse()
 
               
                print('Shortest paath cost from Cient to the Server is ' , Backword_des[goal])
                if astar_path==Backword_des[goal] :
                    return True
                else :
                    return False
            
 
            # for all the neighbors of the current node do
            # لجميع النود المجاوره للنود الحالية يفعلون الاتي 

            for (m, weight) in self.get_neighbors(n):
              # If the current node is not present in both expands and explored, add it to expands.
                # and current node will be in adjacent_nodes
                # إذا كانت النود الحالية غير موجودة في كل من الاكسباند و الاكسبلور ، فقم بإضافتها إلى الاكسباند.
              # وستكون النود الحالية في النود المجاوره
                
                if m not in expands and m not in explored:
                    expands.add(m)
                    adjacent_nodes[m] = n
                    Backword_des[m] = Backword_des[n] + weight
 
                # otherwise, check if it's quicker to first visit n, then m
                # and if it is, update adjacent_nodes data and Backword_des data
                # and if the node was in the explored, move it to expands
                # m اولا ثم n  خلاف ذلك ، تحقق ممن  كان  الأسرع  في الزيارة  
                # Backword_des and adjacent_nodes وإذا كان الأمر كذلك ، فقم بتحديث بيانات    
                # وإذا كانت النود في الاكسبلور ، فانقلها إلى الاكسباند
                
                
                else:
                    if Backword_des[m] > Backword_des[n] + weight:
                        Backword_des[m] = Backword_des[n] + weight
                        adjacent_nodes[m] = n
 
                        if m in explored:
                            explored.remove(m)
                            expands.add(m)
 
            # remove n from the expands, and add it to explored
            # because all of his neighbors were inspected
            #ازاله (ان ) من الاكسباند و اضافتها الى الاكسبلور
            #لأنه تم تفتيش جميع النود المجاوره
            
            expands.remove(n)
            explored.add(n)
 
        print('Path does not exist!')
        return False
    
    def prentpath(self,path):
        for n in path :
            if n=='A':
                print('router A')
            if n=='B':
                print('router B')
            if n=='D':
                print('router D')
            if n=='C':
                print('router C')
            if n=='S':
                print('Client')
            if n=='G':
                print('Server')
```
